```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("train.csv")
data.head()
dataTest = pd.read_csv("test.csv")
dataTest.head()
passengerID = dataTest.PassengerId
```

```python
data.drop(['PassengerId','Name','Ticket','Cabin'],axis=1,inplace=True)
dtype_df = data.dtypes.reset_index()
```

```python
dtype_df.columns = ['count','Type']
```

```python
dtype_df.groupby('Type').aggregate('count')
```

```python
dataTest.drop(['PassengerId','Name','Ticket','Cabin'],axis=1,inplace=True)
dataTest_df = data.dtypes.reset_index()
dataTest_df.columns = ['count','Type']
dataTest_df.groupby('Type').aggregate('count')
```

```python
#处理object数据
object_cols = [x for x in data.columns if data[x].dtype == 'object']
print(object_cols)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le.fit(data['Sex'])
print(le.classes_)
data['Sex'] = le.transform(data['Sex'])
dataTest['Sex'] = le.transform(dataTest['Sex'])
```

```python
Naindex = np.where(data['Embarked'].isna())[0]
for x in Naindex:
    if data.loc[x]['Fare'] == 80.0:
        data.loc[x,'Embarked'] = 'S'
```

```python
#数字化Embark
from sklearn.preprocessing import OneHotEncoder
le = LabelEncoder()
data = data.apply(le.fit_transform)
data.head()
dataTest = dataTest.apply(le.fit_transform)
dataTest.head()
```

```python
from sklearn.model_selection import train_test_split
YLabel = data.Survived
data.drop('Survived',axis=1,inplace=True)
```
